Do wizualizacji wyników w Pythonie używa się pakietu [Matplotlib](https://matplotlib.org/). Działanie tego pakietu prezentuje przykład:

***Zadanie 2.***  
Dla funkcji z zadania pierwszego używając wektora $x \in <-10,10>$ wyrysuj wykres dla argumentów:
1. $x\in<-1,1>$  
2. $x\in<-10,-1>$  
3. $x\in<1,10>$

Używając do tego funkcji: [ylim](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.ylim.html), [xlim](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.xlim.html), [axis](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.axis.html)

```python
# 1.
x = np.linspace(-10,10)
y = x**3 -3*x
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.plot(x, y, label ='x**3 -3*x')

plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.grid()
plt.legend()
plt.show()
# 2.
plt.xlim(-10,-1)
plt.plot(x, y, label ='x**3 -3*x')

plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.grid()
plt.legend()
plt.show()
# 3
plt.axis([1,10,-250,250])
plt.plot(x, y, label ='x**3 -3*x')

plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.grid()
plt.legend()
plt.show()
```

***Komentarz***  
Za pomocą funkcji plt.xlim moge ograniczyć wykres na osi x od xmin do xmax, używając funkcji plt.ylim ograniczam wykres góra-dół, możemy je ze sobą łączyć. Funkcja plt.axis([]) ma tą zalete, że możemy wpisac do niej wektor z interesującymi na ograniczeniami gora-dol oraz lewo-prawo.

***Zadanie 3.*** 
Oblicz ilość ciepła, które wydzieli się podczas hamowania jeżeli opisuje je zależność:  
  
$Q = \cfrac{mv^2}{2}$  
  
jeżeli $m = 2500g$, $v = 60km/h$, wypisz wynik oraz podaj w kilokaloriach i dżulach. Wykreśl wykresy które zobrazują zmianę ilości ciepła w procesie hamowania dla ciała o masie $3000g$ i prędkości $v\in<200,0>km/h$, przedstawiając ciepło hamowania na pierwszym w skali liniowej na drugim w skali logarytmicznej.  

***Wsakzówka:*** Do stworzenia wykresów z osią logarytmiczną skorzystaj z funkcji: [semilogy](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.semilogy.html), [xlim](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xlim.html?highlight=xlim#matplotlib.pyplot.xlim) i [figure](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.figure.html?highlight=figure#matplotlib.pyplot.figure) z pakietu *Matplotlib*. 

```python
# Wstepne obliczenia
m = 2.5                #kg
v = 60 * 1000/3600     #m/s
Q = 0.5 * m * v**2
print("cieplo w Julach:\n{:.5f} J".format(Q))
Q_kcal = Q*0.0002388459
print("cieplo w kilokaloriach:\n{:.5f} kcal".format(Q_kcal))
```

```python
# Wykresowanie wykresów
m_lin=3
v_lin=np.linspace(200,0)                     # jako x zadany
Q_lin=0.5 * m_lin * v_lin**2                 # jako y w funkcji
#SKORZYSTAM Z NOWO POZNANYCH FUNKCJI ZWIAZANYCH Z FUNKCJA FIGURE()
figsize=(16, 5)
fig = plt.figure(figsize=figsize)
ax = fig.add_subplot(1,2,1)

plt.plot(v_lin, Q_lin, label ='0.5 * m * v**2')
plt.xlabel("V - predkosc [m/s]")
plt.ylabel("Q -  cieplo hamowania [J]")
plt.title("Cieplo w funkcji predkosci w skali liniowej")
plt.grid()
plt.legend()


ax = fig.add_subplot(1,2,2)
plt.semilogy(v_lin, Q_lin, label ='0.5 * m * v**2')
plt.xlabel("V - predkosc [m/s]")
plt.ylabel("Q -  cieplo hamowania [J]")
plt.title("Cieplo w funkcji predkosci w skali logarytmicznej")
plt.grid()
plt.legend()
plt.xlim(0,150)

plt.show()
```

***Komentarz***  
Za pomoca funkcji figure tworze okna, robie to za pomoca argumentu figsize, do ktorego zapisuje macierz o wybranych dwoch parametrach: szerokosci i wysokosci. Za pomoca fig.sub_plot(x,y,z) tworze ilosc okienek, x-ilosc wierszy, y-ilosc kolumn, z-indeksowanie poczawszy od lewej strony do prawej.

***Zadanie 6.***   
Innym przydatnym sposobem prezentowania wykresów jest [subplot](https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.subplot.html), czyli wyrysowanie obok siebie kilku wykresów. Zaimplementuj funkcję parallel_plot tak by obok siebie wyrysowane zostały dwa standardowe ploty i użytkownik mógł podać parametry:
- x1 - wektor wartości osi x dla pierwszego wykresu,
- y1 - wektor wartości osi y dla pierwszego wykresu,
- x2 - wektor wartości osi x dla drugiego wykresu,
- y2 - wektor wartości osi x dla drugiego wykresu,
- x1label - opis osi x dla pierwszego wykresu,
- y1label - opis osi y dla pierwszego wykresu,
- x2label - opis osi x dla drugiego wykresu,
- y2label - opis osi y dla drugiego wykresu,
- title - tytuł wykresu,
- orientation - parametr przyjmujący wartość '-' jeżeli subplot ma posiadać dwa wiersze albo '|' jeżeli ma posiadać dwie kolumny.

Jeżeli nie da się wyrysować danych należy zwrócić wartość None.

 

```python
x1=np.linspace(-10, 10)

y1=x1 + 2
label1='x + 2'

x2=x1
y2=x1**2 - 2*np.sin(x1) + 3

x1label='x'
y1label='y'
x2label='x'
y2label='y'
title='Wykres funkcji f(x)'
# I
orientation='czy zadziala'
print(main.parallel_plot(x1,y1,x2,y2,x1label,y1label,x2label,y2label,title,orientation))
# II
orientation='|'
main.parallel_plot(x1,y1,x2,y2,x1label,y1label,x2label,y2label,title,orientation)
plt.show()
# III
orientation='-'
main.parallel_plot(x1,y1,x2,y2,x1label,y1label,x2label,y2label,title,orientation)
plt.show()
```

***Komentarz***  
Jak widac, przekazanie innych argumentow co jest mozliwe spowoduje przekazanie wartości None tak jak w pierwszym przypadku. 

***Zadanie 7.***  
Za pomocą funkcji *parallel\_plot*  i przedstaw na jednym z nich [Spirale logarytmiczną](https://pl.wikipedia.org/wiki/Spirala_logarytmiczna) w szerokim przedziale, a w drugim w okolicy zera.

```python
kat = np.linspace(0,5*np.pi,90)

# Wiekszy przedzial
a=100
b=0.2
r = a*np.exp(b*kat)
x1 = r * np.cos(kat)
y1 = r * np.sin(kat)

label1='x + 2'

# Przedzial w okolicy zera
a=0.01
b=0.2
r = a*np.exp(b*kat)
x2 = r * np.cos(kat)
y2 = r * np.sin(kat)


x1label='x'
y1label='y'
x2label='x'
y2label='y'
title='Spirala logarytmiczna'

orientation='|'
main.parallel_plot(x1,y1,x2,y2,x1label,y1label,x2label,y2label,title,orientation)
plt.show()
```
